# Defining Class for the above function and calling it as BasicSearch

```python
class BasicSearch:
    def __init__(self, amc):
        self.amc = amc
        options = webdriver.ChromeOptions()
        options.add_argument("--headless")  # Enable headless mode
        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

    def open_new_tab_and_search(self, search_query):
        self.driver.execute_script("window.open('about:blank', '_blank');")
        self.driver.switch_to.window(self.driver.window_handles[-1])
        self.driver.get("https://www.google.com")
        search_element = self.driver.find_element(By.NAME, "q")
        search_element.send_keys(search_query)
        search_element.send_keys(Keys.RETURN)

    def extract_relevant_links(self):
        try:
            results = self.driver.find_elements(By.CSS_SELECTOR, 'a[href^="http"][href$=".pdf"], a[href^="http"][href$=".html"]')
            relevant_links = [link.get_attribute("href") for link in results if self.amc in link.get_attribute("href")]

            # Return the first three relevant links
            return relevant_links[:3]

        except Exception as e:
            print(f"Error extracting links: {e}")
            return []

    def print_relevant_links(self, links):
        print("First three relevant links extracted:")
        for i, link in enumerate(links, 1):
            print(f"Link {i}: {link}")

    def search_and_extract_links(self):
        current_year = datetime.now().year
        for year in range(current_year, current_year - 11, -1):
            search_query = f"{self.amc} {year}  Global Macro Outlook filetype:pdf OR filetype:html"
            self.open_new_tab_and_search(search_query)
            relevant_links = self.extract_relevant_links()
            self.print_relevant_links(relevant_links)


# Example usage:
amc = input("Enter AMC name: ")
searcher = BasicSearch(amc)
searcher.search_and_extract_links()
```

# Duplicate link check

```python
# Set up ChromeDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Navigate to a website
driver.get("https://www.google.com")  # Using Google as an example

# Define the list of companies and their names
amc_list = [
    ("KKR", "$180 billion AUM")
]

# Get the current year
current_year = datetime.now().year

# Create a dictionary to store the data
data = {}

# Loop through each AMC
for amc, amc_info in amc_list:
    # Create a dictionary to store links for each year
    amc_data = {}
    # Keep track of found links to avoid duplicates
    found_links = set()

    # Loop through the past 10 years
    for year in range(current_year, current_year - 11, -1):
        search_query = f"{amc} {year} outlook PDF OR HTML"
        try:
            search_input = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.NAME, "q"))
            )
            search_input.clear()
            search_input.send_keys(search_query)
            search_input.send_keys(Keys.RETURN)

            expected_title = search_query
            assert expected_title in driver.title, f"Title mismatch for {search_query}"

            link = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, f"//a[contains(@href,'.pdf') or contains(@href,'.html') and contains(.,'{search_query}')]"))
            )
            href = link.get_attribute("href")

            # Check for duplicate links and modify the link
            if href in found_links:
                # Modify the href by adding a suffix to the last link part before the extension
                parts = href.rsplit('.', 1)
                if parts[0][-2:] == "_":
                    # If already suffixed, just increment the number
                    href = f"{parts[0]}{int(parts[1].split('_')[0]) + 1}.{parts[1]}"
                else:
                    href = f"{parts[0]}_1.{parts[1]}"

            found_links.add(href)
        except Exception as e:
            print(f"Error finding link for {search_query}: {e}")
            href = ""

        amc_data[year] = href
        print(f"Link found for {amc} in {year}: {href}")


    data[amc] = amc_data

# Close the browser
driver.quit()

# Write the data to a CSV file
with open('duplicate_check_KKR.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    header = ['Asset Management Company', *range(current_year, current_year - 11, -1)]
    csv_writer.writerow(header)
    for amc, amc_data in data.items():
        row = [amc]
        for year in range(current_year, current_year - 11, -1):
            row.append(amc_data.get(year, ""))
        csv_writer.writerow(row)

print("Data written to duplicate_check_KKR.csv")
```
